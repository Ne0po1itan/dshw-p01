```python
# %% [markdown]
# # 股票数据分析与可视化
# 分析10只股票的日对数收益率、CAPM模型等

# %% [markdown]
# ## 1. 导入必要的库

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 设置输出文件夹
output_dir = Path('./output')
output_dir.mkdir(exist_ok=True)

# %%
# 股票信息
stocks_info = {
    '600036': {'name': '招商银行', 'industry': '银行'},
    '600104': {'name': '上汽集团', 'industry': '汽车'},
    '000002': {'name': '万科A', 'industry': '房地产'},
    '600048': {'name': '保利地产', 'industry': '房地产'},
    '600519': {'name': '贵州茅台', 'industry': '白酒'},
    '000858': {'name': '五粮液', 'industry': '白酒'},
    '601857': {'name': '中国石油', 'industry': '能源'},
    '601898': {'name': '中煤能源', 'industry': '能源'},
    '600050': {'name': '中国联通', 'industry': '通讯'},
    '600233': {'name': '圆通速递', 'industry': '物流'}
}

# %% [markdown]
# ## 2. 加载数据

# %%
def load_stock_data(code, info):
    """加载单个股票数据"""
    file_path = f'./data/stock/stock_{code}.csv'
    df = pd.read_csv(file_path)
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date').reset_index(drop=True)
    df['code'] = code
    df['name'] = info['name']
    df['industry'] = info['industry']
    # 计算日对数收益率
    df['log_return'] = np.log(df['close'] / df['close'].shift(1))
    return df

# 加载所有股票数据
all_stocks = {}
for code, info in stocks_info.items():
    all_stocks[code] = load_stock_data(code, info)
    print(f"加载完成: {info['name']} ({code}), 数据量: {len(all_stocks[code])}")

# 加载沪深300指数数据
hs300 = pd.read_csv('./data/index/index_000300.csv')
hs300['date'] = pd.to_datetime(hs300['date'])
hs300 = hs300.sort_values('date').reset_index(drop=True)
hs300['log_return'] = np.log(hs300['close'] / hs300['close'].shift(1))
hs300['name'] = '沪深300'
print(f"加载完成: 沪深300, 数据量: {len(hs300)}")

# 加载CPI数据
cpi = pd.read_csv('./data/macro/macro_cpi.csv')
# 处理CPI日期格式（2026年02月份 -> 2026-02）
cpi['date_str'] = cpi['date'].astype(str)
# 使用str.extract提取年月
cpi_extracted = cpi['date_str'].str.extract(r'(\d{4})年(\d{2})月份')
cpi['year'] = cpi_extracted[0]
cpi['month'] = cpi_extracted[1]
cpi['date'] = pd.to_datetime(cpi['year'] + '-' + cpi['month'])
cpi = cpi.drop(['date_str', 'year', 'month'], axis=1)
print(f"加载完成: CPI数据, 数据量: {len(cpi)}")

# %% [markdown]
# ## 3. 描述性统计

# %%
def annualized_stats(returns):
    """计算年化统计指标"""
    # 假设交易日252天
    ret_clean = returns.dropna()
    if len(ret_clean) == 0:
        return np.nan, np.nan, np.nan, np.nan
    
    # 年化均值
    ann_mean = ret_clean.mean() * 252
    
    # 年化波动率
    ann_vol = ret_clean.std() * np.sqrt(252)
    
    # 偏度
    skewness = ret_clean.skew()
    
    # 峰度
    kurtosis = ret_clean.kurtosis()
    
    return ann_mean, ann_vol, skewness, kurtosis

def max_drawdown(price_series):
    """计算最大回撤"""
    cummax = price_series.expanding().max()
    drawdown = (price_series - cummax) / cummax
    return drawdown.min()

# 计算描述性统计
desc_stats = []
for code, info in stocks_info.items():
    df = all_stocks[code]
    ret = df['log_return'].dropna()
    
    ann_mean, ann_vol, skew, kurt = annualized_stats(ret)
    mdd = max_drawdown(df['close'])
    
    desc_stats.append({
        '股票': info['name'],
        '行业': info['industry'],
        '年化均值': ann_mean,
        '年化波动率': ann_vol,
        '偏度': skew,
        '峰度': kurt,
        '最大回撤': mdd
    })

desc_df = pd.DataFrame(desc_stats)
print("描述性统计结果：")
print(desc_df.to_string(index=False))

# %% [markdown]
# ## 4. 归一化收盘价走势图

# %%
# 计算归一化收盘价（以2020-01-01为基准=1）
start_date = pd.to_datetime('2020-01-01')

plt.figure(figsize=(14, 8))

# 定义行业颜色
industries = list(set([info['industry'] for info in stocks_info.values()]))
colors = plt.cm.tab10(np.linspace(0, 1, len(industries)))
industry_color = dict(zip(industries, colors))

# 绘制股票
for code, info in stocks_info.items():
    df = all_stocks[code]
    # 筛选2020-01-01之后的数据
    df_plot = df[df['date'] >= start_date].copy()
    if len(df_plot) > 0:
        # 归一化
        base_price = df_plot['close'].iloc[0]
        df_plot['norm_close'] = df_plot['close'] / base_price
        plt.plot(df_plot['date'], df_plot['norm_close'], 
                label=info['name'], color=industry_color[info['industry']], 
                linewidth=1.5, alpha=0.7)

# 绘制沪深300
hs300_plot = hs300[hs300['date'] >= start_date].copy()
if len(hs300_plot) > 0:
    base_hs300 = hs300_plot['close'].iloc[0]
    hs300_plot['norm_close'] = hs300_plot['close'] / base_hs300
    plt.plot(hs300_plot['date'], hs300_plot['norm_close'], 
            label='沪深300', color='black', linewidth=2.5, linestyle='--')

plt.xlabel('日期', fontsize=12)
plt.ylabel('归一化收盘价 (2020-01-01 = 1)', fontsize=12)
plt.title('10只股票与沪深300归一化收盘价走势对比', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(output_dir / 'normalized_price.png', dpi=150, bbox_inches='tight')
plt.show()
print("已保存: normalized_price.png")

# %% [markdown]
# ## 5. 日收益率分布图

# %%
# 2行×5列的分面直方图
fig, axes = plt.subplots(2, 5, figsize=(20, 10))
axes = axes.flatten()

for idx, (code, info) in enumerate(stocks_info.items()):
    df = all_stocks[code]
    ret = df['log_return'].dropna()
    
    # 绘制直方图
    axes[idx].hist(ret, bins=50, density=True, alpha=0.6, color='steelblue', edgecolor='black')
    
    # 叠加正态分布曲线
    x = np.linspace(ret.min(), ret.max(), 100)
    mu, std = ret.mean(), ret.std()
    axes[idx].plot(x, stats.norm.pdf(x, mu, std), 'r-', linewidth=2, label='正态分布')
    
    # 标注均值和标准差
    axes[idx].axvline(mu, color='green', linestyle='--', linewidth=1, alpha=0.7)
    axes[idx].text(0.05, 0.95, f'均值: {mu:.4f}\n标准差: {std:.4f}', 
                   transform=axes[idx].transAxes, verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    axes[idx].set_title(f'{info["name"]} ({info["industry"]})', fontsize=10)
    axes[idx].set_xlabel('日对数收益率')
    axes[idx].set_ylabel('密度')
    axes[idx].legend(fontsize=8)
    axes[idx].grid(True, alpha=0.3)

plt.suptitle('10只股票日收益率分布直方图', fontsize=16, y=1.02)
plt.tight_layout()
plt.savefig(output_dir / 'return_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("已保存: return_distribution.png")

# %% [markdown]
# ## 6. 收益率相关系数热力图

# %%
# 构建收益率矩阵
returns_matrix = pd.DataFrame()
for code, info in stocks_info.items():
    df = all_stocks[code]
    returns_matrix[info['name']] = df.set_index('date')['log_return']

# 按行业排序
sort_order = []
for industry in ['银行', '汽车', '房地产', '白酒', '能源', '通讯', '物流']:
    for code, info in stocks_info.items():
        if info['industry'] == industry:
            sort_order.append(info['name'])
returns_matrix = returns_matrix[sort_order]

# 计算相关系数
corr_matrix = returns_matrix.corr()

# 绘制热力图
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='RdYlBu_r', 
            center=0, square=True, linewidths=0.5, 
            cbar_kws={"shrink": 0.8}, mask=mask)

plt.title('10只股票日收益率相关系数矩阵热力图', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig(output_dir / 'correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("已保存: correlation_heatmap.png")

# 讨论同行业相关性
print("\n同行业内部平均相关性：")
industries_corr = {}
for industry in set([info['industry'] for code, info in stocks_info.items()]):
    industry_stocks = [info['name'] for code, info in stocks_info.items() if info['industry'] == industry]
    if len(industry_stocks) > 1:
        sub_corr = corr_matrix.loc[industry_stocks, industry_stocks]
        # 取上三角平均值（不包括对角线）
        avg_corr = sub_corr.values[np.triu_indices_from(sub_corr.values, k=1)].mean()
        industries_corr[industry] = avg_corr
        print(f"  {industry}: {avg_corr:.4f}")

# 计算跨行业平均相关性
all_corr_values = corr_matrix.values[np.triu_indices_from(corr_matrix.values, k=1)]
print(f"\n跨行业平均相关性: {all_corr_values.mean():.4f}")
print("\n结论：同行业内部相关性普遍高于跨行业平均相关性，表明行业因素对股价走势有显著影响。")

# %% [markdown]
# ## 7. CPI与股市关系图

# %%
# 计算沪深300月度收益率
hs300_monthly = hs300.copy()
hs300_monthly['year_month'] = hs300_monthly['date'].dt.to_period('M')
monthly_returns = hs300_monthly.groupby('year_month')['log_return'].apply(lambda x: x.sum())
monthly_returns = monthly_returns.reset_index()
monthly_returns['date'] = monthly_returns['year_month'].dt.to_timestamp()
monthly_returns = monthly_returns.dropna()

# 合并CPI数据
cpi_merged = cpi.copy()
cpi_merged['year_month'] = cpi_merged['date'].dt.to_period('M')
merged = pd.merge(monthly_returns, cpi_merged, on='year_month', how='inner')
merged = merged.dropna()

# 计算相关系数
pearson_corr, p_value = stats.pearsonr(merged['log_return'], merged['cpi_yoy'])

# 绘制散点图
plt.figure(figsize=(10, 8))
plt.scatter(merged['cpi_yoy'], merged['log_return'], alpha=0.6, s=50, color='steelblue')

# 线性拟合
z = np.polyfit(merged['cpi_yoy'], merged['log_return'], 1)
p = np.poly1d(z)
x_line = np.linspace(merged['cpi_yoy'].min(), merged['cpi_yoy'].max(), 100)
plt.plot(x_line, p(x_line), 'r-', linewidth=2, label=f'线性拟合: y={z[0]:.4f}x+{z[1]:.4f}')

plt.xlabel('CPI同比增速 (%)', fontsize=12)
plt.ylabel('沪深300月度收益率', fontsize=12)
plt.title(f'CPI与沪深300月度收益率关系图\nPearson相关系数: {pearson_corr:.4f} (p={p_value:.4f})', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig(output_dir / 'cpi_vs_hs300.png', dpi=150, bbox_inches='tight')
plt.show()
print("已保存: cpi_vs_hs300.png")

print(f"\nPearson相关系数: {pearson_corr:.4f}, p值: {p_value:.4f}")
print("\n经济含义讨论：")
if pearson_corr > 0:
    print("CPI与股市收益率呈正相关关系。")
    print("可能原因：适度通胀反映经济增长，企业盈利改善，推动股市上涨。")
else:
    print("CPI与股市收益率呈负相关关系。")
    print("可能原因：")
    print("1. 高通胀预期可能导致央行收紧货币政策，对股市形成压力")
    print("2. 通胀上升侵蚀企业利润，降低股票估值")
    print("3. 投资者可能转向抗通胀资产，减少股票配置")

# %% [markdown]
# ## 8. CAPM模型估计

# %%
# 准备市场收益率数据（沪深300）
market_returns = hs300.set_index('date')['log_return']

# 设置无风险利率（年化2%）
rf_annual = 0.02  # 2%年化无风险利率
# 转换为日收益率（假设252个交易日）
rf_daily = (1 + rf_annual) ** (1/252) - 1
print(f"日度无风险利率: {rf_daily:.6f} ({rf_daily*100:.4f}%)")

# 估计CAPM模型
capm_results = []

for code, info in stocks_info.items():
    df = all_stocks[code]
    # 对齐数据
    stock_returns = df.set_index('date')['log_return']
    merged_returns = pd.merge(stock_returns, market_returns, 
                              left_index=True, right_index=True, suffixes=('_stock', '_market'))
    merged_returns = merged_returns.dropna()
    
    # 超额收益（使用日度无风险利率）
    excess_stock = merged_returns['log_return_stock'] - rf_daily
    excess_market = merged_returns['log_return_market'] - rf_daily
    
    if len(excess_stock) > 0:
        # OLS回归
        X = sm.add_constant(excess_market)
        model = sm.OLS(excess_stock, X).fit()
        
        alpha = model.params['const']
        beta = model.params['log_return_market']
        p_value = model.pvalues['const']
        r_squared = model.rsquared
        
        # 计算Beta的95%置信区间
        beta_ci = model.conf_int().loc['log_return_market']
        
        capm_results.append({
            '股票': info['name'],
            '行业': info['industry'],
            'α': alpha,
            'α_p值': p_value,
            'α_显著': p_value < 0.05,
            'β': beta,
            'β_lower_95': beta_ci[0],
            'β_upper_95': beta_ci[1],
            'R²': r_squared
        })

capm_df = pd.DataFrame(capm_results)
print("CAPM模型估计结果（无风险利率=2%年化）：")
print(capm_df[['股票', '行业', 'α', 'α_p值', 'β', 'R²']].to_string(index=False))

# %% [markdown]
# ## 9. Beta系数点图

# %%
# 绘制Beta系数点图
fig, ax = plt.subplots(figsize=(12, 8))

# 按行业分组着色
industries_list = list(capm_df['行业'].unique())
colors_beta = plt.cm.tab10(np.linspace(0, 1, len(industries_list)))
industry_color_beta = dict(zip(industries_list, colors_beta))

# 为每个股票绘制点图和误差棒
for idx, row in capm_df.iterrows():
    color = industry_color_beta[row['行业']]
    y_pos = idx
    
    # 绘制点
    ax.scatter(row['β'], y_pos, color=color, s=100, zorder=5, edgecolors='black', linewidth=1)
    # 绘制误差棒
    ax.errorbar(row['β'], y_pos, 
                xerr=[[row['β'] - row['β_lower_95']], [row['β_upper_95'] - row['β']]],
                fmt='none', color=color, capsize=5, capthick=2, elinewidth=2, zorder=3)

# 设置y轴标签
ax.set_yticks(range(len(capm_df)))
ax.set_yticklabels(capm_df['股票'])
ax.set_xlabel('Beta系数', fontsize=12)
ax.set_ylabel('股票名称', fontsize=12)
ax.set_title('CAPM模型Beta系数估计（带95%置信区间）', fontsize=14)

# 添加β=1参考线
ax.axvline(x=1, color='red', linestyle='--', linewidth=2, alpha=0.7, label='β=1 (市场风险)')

# 添加网格
ax.grid(True, alpha=0.3, axis='x')

# 添加图例
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', 
                              markerfacecolor=industry_color_beta[industry], 
                              markersize=10, label=industry) 
                   for industry in industries_list]
legend_elements.append(plt.Line2D([0], [0], color='red', linestyle='--', linewidth=2, label='β=1'))
ax.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig(output_dir / 'beta_dot_plot.png', dpi=150, bbox_inches='tight')
plt.show()
print("已保存: beta_dot_plot.png")

# %% [markdown]
# ## 10. CAPM模型结果分析

# %%
print("\n" + "="*80)
print("CAPM模型结果分析")
print("="*80)

# 问题1: β>1的股票
beta_gt_1 = capm_df[capm_df['β'] > 1]
print("\n【问题1】β > 1的股票：")
if len(beta_gt_1) > 0:
    for _, row in beta_gt_1.iterrows():
        print(f"  {row['股票']} ({row['行业']}): β = {row['β']:.4f}")
else:
    print("  无股票β > 1")

print("\n与'周期性 vs 防御性'行业分类的吻合性分析：")
print("  - β > 1表示股票波动性大于市场，属于'进攻型'或'周期性'行业")
print("  - β < 1表示股票波动性小于市场，属于'防御型'行业")
print("  - 从结果来看：")
if len(beta_gt_1) > 0:
    for _, row in beta_gt_1.iterrows():
        if row['行业'] in ['汽车', '房地产', '能源']:
            print(f"    ✓ {row['股票']}({row['行业']})符合周期性行业特征")
        elif row['行业'] in ['银行', '白酒', '通讯', '物流']:
            print(f"    ? {row['股票']}({row['行业']})通常被视为防御性或弱周期，但β>1需关注具体原因")
else:
    print("  所有股票β均小于1，符合防御性特征或数据期间市场波动较大")

# 问题2: Alpha显著性
alpha_significant = capm_df[capm_df['α_显著'] == True]
print("\n【问题2】α显著异于零的股票（p值 < 0.05）：")
if len(alpha_significant) > 0:
    for _, row in alpha_significant.iterrows():
        sign = "正" if row['α'] > 0 else "负"
        print(f"  {row['股票']} ({row['行业']}): α = {row['α']:.6f} ({sign}值), p值 = {row['α_p值']:.4f}")
else:
    print("  无股票α显著异于零")

print("\nα显著的经济含义：")
print("  - α（Alpha）代表超额收益，即股票实际收益超出CAPM模型预测的部分")
print("  - α显著为正：股票具有超额收益，可能由于公司特有优势、市场定价错误或风险未完全被模型捕捉")
print("  - α显著为负：股票表现低于风险调整后的预期，可能存在管理问题或负面因素")
print("  - α不显著：股票收益可以很好地被市场风险解释，没有显著超额收益")

# 问题3: R²最高和最低
max_r2 = capm_df.loc[capm_df['R²'].idxmax()]
min_r2 = capm_df.loc[capm_df['R²'].idxmin()]
print(f"\n【问题3】R²最高的股票：{max_r2['股票']} ({max_r2['行业']}), R² = {max_r2['R²']:.4f}")
print(f"        R²最低的股票：{min_r2['股票']} ({min_r2['行业']}), R² = {min_r2['R²']:.4f}")

print("\nR²差异的解释：")
print(f"  - {max_r2['股票']}的R² = {max_r2['R²']:.4f}，说明其收益变动的{max_r2['R²']*100:.1f}%可由市场收益解释")
print(f"  - {min_r2['股票']}的R² = {min_r2['R²']:.4f}，说明其收益变动的{min_r2['R²']*100:.1f}%可由市场收益解释")
print("  造成差异的可能原因：")
print("  1. 行业属性：周期性行业（如汽车、房地产）通常与市场相关性更高")
print("  2. 公司规模：大盘股受市场整体影响更大，小盘股可能有更多特异性风险")
print("  3. 业务模式：业务多元化的公司可能受市场影响较小")
print("  4. 监管环境：受政策影响大的行业（如能源）可能与市场相关性较低")
print("  5. 投资者结构：机构持股比例高的股票更可能跟随市场走势")

# %%
print("\n分析完成！所有图表已保存至output文件夹。")
```

### CAPM 模型估计
1. β > 1的股票：

与'周期性 vs 防御性'行业分类的吻合性分析：
  - β > 1表示股票波动性大于市场，属于'进攻型'或'周期性'行业
  - β < 1表示股票波动性小于市场，属于'防御型'行业
  - 从结果来看：   
    - ✓ 万科A(房地产)符合周期性行业特征，房地产政策往往在一段时间内呈单边趋势（如持续从严或持续宽松），这种政策惯性容易催生板块的趋势性行情，为进攻提供宏观环境。
    - ? 五粮液(白酒)通常被视为兼具进攻与防御属性的“双栖型”行业，白酒是典型的顺周期消费品，其需求与商务活动、居民收入预期高度相关。当宏观经济向好时，白酒板块往往能获得超越大盘的涨幅。白酒承载文化和情感属性，能够获取高附加值，龙头品牌拥有深厚的“护城河”，在行业下行期具备抗跌能力。

  - α显著的经济含义：股票表现低于风险调整后的预期，可能存在管理问题或负面因素

  - 五粮液的R² = 0.4989，说明其收益变动的49.9%可由市场收益解释
  - 中煤能源的R² = 0.0743，说明其收益变动的7.4%可由市场收益解释
  - 五粮液的R²最高（约0.50），是因为它作为白酒龙头和沪深300核心成分股，与宏观经济、消费升级及机构资金流向高度相关，其股价波动中约一半可由市场整体走势解释；而中煤能源的R²极低（约0.07），则是因为煤炭股主要受煤价、供需政策、安全生产等行业特有因素驱动，与沪深300所代表的大盘蓝筹行情关联很弱，因此市场波动只能解释其股价变动的极小部分。
